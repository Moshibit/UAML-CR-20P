# Importamos librerías a utilizar

```python
#Para crear y análizar redes
import networkx as nx
#Para dibujar redes y graficar funciones
import matplotlib.pyplot as plt
```

# Creación de Distintas redes

Creamos una red completa de 100 nodos

```python
K_100 = nx.complete_graph(100)
```

Creamos una red aleatoria de 100 nodos y probabilidad de conexión de 0.06 - (Algoritmo de Gilbert con parámteros n=100, p=0.06)

```python
rnd = nx.erdos_renyi_graph(100, 0.06)
```

Creamos una red de mundo pequeño con 100 nodos y probabilidad de reconexión de p=0.3 (Algoritmo de Watts-Strotgatz con parámetros n=100, k=4, p=0.3)

```python
sw = nx.connected_watts_strogatz_graph(100, 4, 0.3)
```

Creamos una red de escala libre con 100 nodos y 2 enlaces por cada nuevo nodo agregado a la red (Algoritmo Barabsi-Albert n=100, 2)

```python
sf = nx.barabasi_albert_graph(100, 2)
```

# Información de cada red
## Red Completa
Calculamos la información

```python
# funciones de la grafica creada por networkX
numero_nodos_k = K_100.number_of_nodes()
numero_enlaces_k = K_100.number_of_edges()
# funciones de networkX
diameter_k = nx.diameter(K_100)
densidad_k = nx.density(K_100)
```

Imprimimos la información

```python
print ("Nodos: " + str(numero_nodos_k))
print ("Enlaces: " + str(numero_enlaces_k))
print ("Diametro: " + str(diameter_k))
print ("Densidad: " + str(densidad_k))
```

# Distribución de Grados
Calculamos y gráficamos la distribución de grados de cada red

```python
# Red Completa
dist_k = nx.degree_histogram(K_100)
print(dist_k)
print(len(dist_k))
print(dist_k[0]) #0 nodos de grado 0 (dist_k[0]=0)
print(dist_k[56]) #0 nodos de grado 57 (dist_k[56]=0)
print(dist_k[99])#99 nodos de grado 100 (dist_k[99]=100)
```

```python
# Red Aleatoria
dist_rnd = nx.degree_histogram(rnd)
print(dist_rnd)
print(len(dist_rnd))
print(dist_rnd[0])
print(dist_rnd[6])
print(dist_rnd[11])
```

```python
# Red de Mundo Pequeño
dist_sw = nx.degree_histogram(sw)
print(dist_sw)
print(len(dist_sw))
print(dist_sw[0])
print(dist_sw[4])
print(dist_sw[7])
```

```python
# Red de Escala Libre
dist_sf = nx.degree_histogram(sf)
print(dist_sf)
print(len(dist_sf))
print(dist_sf[0])
print(dist_sf[11])
print(dist_sf[21])
```

Graficamos las distribuciones

```python
# Las 4 gráficas muestran la distribución de la red completa. Se debe modificar este código 
# para que se muestre la distribución las redes complta, aleatoria, mundo pequeño y escala libre
fig, axs = plt.subplots(2, 2)
fig.suptitle("Distribución de grados", y=1.05)
axs[0, 0].plot(dist_k)
axs[0, 0].set_title('Red Completa')
axs[0, 1].plot(dist_rnd)
axs[0, 1].set_title('Red Aleatoria')
axs[1, 0].plot(dist_sw)
axs[1, 0].set_title('Red de Mundo Pequeño')
axs[1, 1].plot(dist_sf)
axs[1, 1].set_title('Red de Escala libre')

for ax in axs.flat:
    ax.set(xlabel='Grado', ylabel='# Nodos')

fig.tight_layout()
```
